```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('dataset.csv')
print(df.shape)
df.head()
```

```python
df = df.dropna()
df = df.drop_duplicates()
print(df.shape)
```

```python
top_genres = df.groupby('track_genre')['popularity'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
top_genres.plot(kind='bar', color='steelblue')
plt.title('Top 10 Most Popular Music Genres on Spotify')
plt.xlabel('Genre')
plt.ylabel('Average Popularity Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('top_genres.png')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['energy'], df['popularity'], alpha=0.3, color='coral')
plt.title('Energy vs Popularity')
plt.xlabel('Energy')
plt.ylabel('Popularity')
plt.tight_layout()
plt.savefig('energy_vs_popularity.png')
plt.show()
```

```python
top_artists = df.groupby('artists')['popularity'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
top_artists.plot(kind='barh', color='mediumseagreen')
plt.title('Top 10 Most Popular Artists')
plt.xlabel('Average Popularity Score')
plt.tight_layout()
plt.savefig('top_artists.png')
plt.show()
```

```python
# Summary of findings
print("=== SPOTIFY ANALYSIS SUMMARY ===")
print(f"Total songs analyzed: {len(df):,}")
print(f"\nTop genre: {top_genres.index[0]} (avg popularity: {top_genres.values[0]:.1f})")
print(f"Top artist: {top_artists.index[0]} (avg popularity: {top_artists.values[0]:.1f})")
print("\nKey insight: Higher energy songs tend to score higher in popularity")
print("Charts saved: top_genres.png, energy_vs_popularity.png, top_artists.png")
```
